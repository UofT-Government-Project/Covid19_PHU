```python
# Importing dependencies
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Reading file source
file_path = Path('Datasource/sample_covid_dataset.csv')
cases_df = pd.read_csv(file_path)
cases_df.head()
```

```python
# checking details of dataframe
cases_df.info()
```

```python
# dropping null values
cases_df.dropna()
```

```python
# getting column names
cases_df.columns
```

```python
# dropping unneccessary columns
PHU_cases_df = cases_df.drop(columns=['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1','Unnamed: 0.1.1.1','_id','Row_ID',
                'Accurate_Episode_Date','Test_Reported_Date','Specimen_Date','Case_AcquisitionInfo','Reporting_PHU_Address',
                'Reporting_PHU_City','Reporting_PHU_Postal_Code','Reporting_PHU_Website'])
PHU_cases_df
```

```python
# renaming columns 
PHU_cases_df = PHU_cases_df.rename(columns={'Case_Reported_Date':'Case_Date', 'Client_Gender':'Gender', 'Outcome1':'Status',
                                   'Reporting_PHU_ID':'PHU_ID', 'Reporting_PHU':'PHU','Reporting_PHU_Latitude':'PHU_Lat',
                                   'Reporting_PHU_Longitude':'PHU_Long'}, inplace=False)
PHU_cases_df
```

```python
# extracting month, week, year from "Case_Date"
PHU_cases_df['Case_Month'] = pd.DatetimeIndex(PHU_cases_df['Case_Date']).month
PHU_cases_df['Case_Year']= pd.DatetimeIndex(PHU_cases_df['Case_Date']).year
PHU_cases_df
```

```python
# dropping "Case_Date" column
PHU_cases_df.drop(columns=['Case_Date'])
```

```python
# saving cleaned dataset
output_file = Path('Datasource/clean_sample_covid_dataset.csv')
PHU_cases_df.to_csv(output_file, index=False)
```
