```python
# Importing dependencies
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Reading file source
file_path = Path('Datasource/sample_covid_dataset.csv')
cases_df = pd.read_csv(file_path)
cases_df.head()
```

```python
# checking details of dataframe
cases_df.info()
```

```python
# getting column names
cases_df.columns
```

```python
# dropping unneccessary columns
PHU_cases_df = cases_df.drop(columns=['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1','Unnamed: 0.1.1.1','_id',
                'Accurate_Episode_Date','Test_Reported_Date','Specimen_Date','Case_AcquisitionInfo','Reporting_PHU_Address',
                'Reporting_PHU_City','Reporting_PHU_Postal_Code','Reporting_PHU_Website'])
PHU_cases_df
```

```python
# dropping duplicates if any
clean_df = PHU_cases_df.drop_duplicates()
clean_df
```

```python
# dropping any null values
clean_df.dropna()
```

```python
# renaming columns 
clean_PHU_data = clean_df.rename(columns={'Case_Reported_Date':'Case_Date', 'Client_Gender':'Gender', 'Outcome1':'Status',
                                   'Reporting_PHU_ID':'PHU_ID', 'Reporting_PHU':'PHU','Reporting_PHU_Latitude':'PHU_Lat',
                                   'Reporting_PHU_Longitude':'PHU_Long'}, inplace=False)
clean_PHU_data
```

```python
# extracting month, week, year from "Case_Date"
clean_PHU_data['Case_Month'] = pd.DatetimeIndex(clean_PHU_data['Case_Date']).month
clean_PHU_data['Case_Year']= pd.DatetimeIndex(clean_PHU_data['Case_Date']).year
clean_PHU_data
```

```python
# dropping "Case_Date" column
revised_clean_data = clean_PHU_data.drop(columns=['Case_Date'])
revised_clean_data
```

```python
# saving cleaned dataset
output_file = Path('Datasource/revised_clean_dataset.csv')
revised_clean_data.to_csv(output_file, index=False)
```
